```python
#Histogram of runs scored(Batting)
plt.figure(figsize=(10,5))
sns.histplot(bat['Runs Scored'],bins=20,kde=True)
plt.title("Distribution of Runs scored by players")
plt.xlabel("Runs Scored")
plt.ylabel("Frequency")
plt.show()
```

```python
#Histogram of wickets taken(Bowling)
plt.figure(figsize=(10,5))
sns.histplot(bowl['Wickets'],bins=15,kde=True,color='red')
plt.title("Distribution of Wickets Taken by players")
plt.xlabel("Wickets")
plt.ylabel("Frequency")
plt.show()
```

```python
#correlation heatmap for batting
plt.figure(figsize=(10,6))
sns.heatmap(bat.select_dtypes(include=['number']).corr(),annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Correlation Heatmap - Batting")
plt.show()
```

```python
#correlation heatmap for bowling
plt.figure(figsize=(10,6))
sns.heatmap(bowl.select_dtypes(include=['number']).corr(),annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Correlation Heatmap - Batting")
plt.show()
```

```python
#identifying top 10 batsmen based on runs scored
top_scorers=bat.nlargest(10,'Runs Scored')

#plot bar chart for total runs
fig,ax1=plt.subplots(figsize=(12,6))
sns.barplot(x="Player",y="Runs Scored",data=top_scorers,ax=ax1,palette="Blues_r")
ax1.set_ylabel("Total Runs",color="blue")
ax1.set_xlabel("Players")
ax1.set_title("Top 10 Run Scorers in ICC Men's T20 World Cup 2024")
ax1.set_xticklabels(top_scorers["Player"],rotation=45,ha="right")


plt.show()
```

```python
#identifying top 10 bowlers based on runs scored
top_scorers=bowl.nlargest(10,'Wickets')

#plot bar chart for wickets
fig,ax1=plt.subplots(figsize=(12,6))
sns.barplot(x="Player",y="Wickets",data=top_scorers,ax=ax1,palette="Reds_r")
ax1.set_ylabel("Total Wickets",color="red")
ax1.set_xlabel("Players")
ax1.set_title("Top 10 Wicket-takers in ICC Men's T20 World Cup 2024")
ax1.set_xticklabels(top_scorers["Player"],rotation=45,ha="right")


plt.show()
```

```python
#create a new column to see if the toss winner also wins matches
t20["Toss Impact"]=t20["Toss Winning"]==t20["Winners"]

#count matches where toss winner won the match
toss_win_count=t20["Toss Impact"].value_counts()

#toss impact analysis
plt.figure(figsize=(8,5))
sns.barplot(x=toss_win_count.index,y=toss_win_count.values,palette=["red","green"])
plt.xticks([0,1],["Lost Match","Won Match"])
plt.ylabel("Number of matches")
plt.xlabel("Did toss winner won the match?")
plt.title("Impact of Toss on Match Results")
plt.show()
```

```python
#toss decision analysis
toss_decision=t20.groupby("Toss Decision")["Winners"].count()
plt.figure(figsize=(8,5))
sns.barplot(x=toss_decision.index,y=toss_decision.values,palette="coolwarm")
plt.ylabel("Num of Matches Won")
plt.xlabel("Toss Decision")
plt.title("Match Wins Based on Toss Decision(Bat First vs Chase)")
plt.show()
```

```python
#matches count per venue
venue_count=t20["Venue Stadium"].value_counts()

#plot matches per venue
plt.figure(figsize=(12,6))
sns.barplot(x=venue_count.index,y=venue_count.values,palette="viridis")
plt.xticks(rotation=90)
plt.xlabel("Venue")
plt.ylabel("Num of Matches")
plt.title("Num of Matches Played at Each Venue")
plt.show()
```

```python
# Average winning margin per venue
venue_margin = t20.groupby("Venue Stadium")["Winning Margin"].mean().sort_values(ascending=False)

# Plot: Average winning margin per venue
plt.figure(figsize=(12,6))
sns.barplot(x=venue_margin.index, y=venue_margin.values, palette="magma")
plt.xticks(rotation=90)
plt.xlabel("Venue")
plt.ylabel("Average Winning Margin")
plt.title("Average Winning Margin at Each Venue")
plt.show()
```

```python
# Toss decision impact at different venues
venue_toss = t20.groupby(["Venue Stadium", "Toss Decision"])["Winners"].count().unstack()

# Plot: Toss Decision by Venue
venue_toss.plot(kind="bar", stacked=True, figsize=(14,6), colormap="coolwarm")
plt.xlabel("Venue")
plt.ylabel("Match Wins")
plt.title("Match Wins Based on Toss Decision at Different Venues")
plt.xticks(rotation=90)
plt.legend(title="Toss Decision")
plt.show()
```

```python
#count of total wins per team
team_wins=t20["Winners"].value_counts()

#plot match wins for team
plt.figure(figsize=(12,6))
sns.barplot(x=team_wins.index,y=team_wins.values,palette="Blues_r")
plt.xlabel("Team")
plt.ylabel("Num of Matches Won")
plt.title("Total Matches Won by Each Team")
plt.xticks(rotation=45)
plt.show()
```

```python
#average winning margin per team
team_margin=t20.groupby("Winners")["Winning Margin"].mean().sort_values(ascending=False)

#plotting the same
plt.figure(figsize=(12,6))
sns.barplot(x=team_margin.index,y=team_margin.values,palette="Reds_r")
plt.xlabel("Team")
plt.ylabel("Average Winning Margin")
plt.title("Average Winning Margin Per Team")
plt.xticks(rotation=45)
plt.show()
```

```python
# Check the distribution of Winning Margin and analyze match outcomes
plt.figure(figsize=(12, 6))
# Ensure the 'Winning Margin' column is numeric
t20['Winning Margin'] = pd.to_numeric(t20['Winning Margin'], errors='coerce')

# Filter rows by won by (runs and wickets)
runs_margin = t20[t20['Won by'] == 'Runs']['Winning Margin']
wickets_margin = t20[t20['Won by'] == 'Wickets']['Winning Margin']

# Plot the distribution of winning margins by runs
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
sns.histplot(runs_margin.dropna(), bins=20, kde=True, color='blue')
plt.title('Distribution of Winning Margin (by Runs)')
plt.xlabel('Winning Margin (Runs)')
plt.ylabel('Frequency')
```

```python
#plot the distribution of winning margins by wickets
plt.subplot(2,1,2)
sns.histplot(wickets_margin.dropna(),bins=20,kde=True,color='green')
plt.title("Distribution of Winning Margin(by Wickets)")
plt.xlabel("Winning Margin(wickets)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Assuming 'Match Referee' is the column with the match referees (replace with the actual column name if different)
# Count the occurrences of each match referee
match_referee_counts = t20['Match Referee'].value_counts()

# Select the top 5 match referees
top_5_match_referees = match_referee_counts.nlargest(5)

# Print the top 5 match referees
print("\nTop 5 Match Referees:")
print(top_5_match_referees)

# Plot the top 5 match referees
plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_match_referees.values, y=top_5_match_referees.index, palette='copper')

# Adding titles and labels
plt.title('Top 5 Match Referees in ICC T20 World Cup')
plt.xlabel('Number of Matches')
plt.ylabel('Match Referee')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#filter data for indian players
indian_batters = bat[bat["Player"].str.contains("(IND)", regex=False)]
indian_bowlers = bowl[bowl["Player"].str.contains("(IND)", regex=False)]
indian_matches=t20[t20["Winners"]=="India"]

#top indian batsman analysis
#aggregating total runs and highest score
top_batsman=indian_batters.groupby("Player").agg({"Runs Scored":"sum","Batting Strike Rate": "mean", "Highest Innings Score":"max"})
top_batsman=top_batsman.sort_values(by="Runs Scored",ascending=False)

#plot top indian batsman
plt.figure(figsize=(10,5))
sns.barplot(x=top_batsman["Runs Scored"],y=top_batsman.index,palette="Blues_r")
plt.xlabel("Total Runs")
plt.ylabel("Player")
plt.title("Top Indian Batsman in T20 World Cup 2024")
plt.show()
```

```python
indian_bowlers = bowl[bowl["Player"].str.contains("(IND)", regex=False)]

#top indian bowlers analysis
#aggregating wickets taken and economy rate
top_batsman=indian_bowlers.groupby("Player").agg({"Wickets":"sum","Economy Rate": "mean"})
top_batsman=top_batsman.sort_values(by="Wickets",ascending=False)

#plot top indian batsman
plt.figure(figsize=(10,5))
sns.barplot(x=top_batsman["Wickets"],y=top_batsman.index,palette="Reds_r")
plt.xlabel("Wickets Taken")
plt.ylabel("Player")
plt.title("Top Indian Bowlers in T20 World Cup 2024")
plt.show()
```

```python
# Count Player of the Match awards for Indian players
india_pom = indian_matches["Player Of The Match"].value_counts()

# Plot: Most Impactful Players (POTM Awards)
plt.figure(figsize=(10,5))
sns.barplot(y=india_pom.index, x=india_pom.values, palette="Purples_r")
plt.xlabel("Number of POTM Awards")
plt.ylabel("Player")
plt.title("Most Impactful Indian Players in T20 World Cup 2024")
plt.show()
```

```python
# Rename columns for clarity
runs_scored = t20[['1st Team', 'First Innings Score']].rename(columns={'1st Team': 'Team', 'First Innings Score': 'Runs'})

# Ensure 'Runs' column is numeric
runs_scored['Runs'] = pd.to_numeric(runs_scored['Runs'], errors='coerce')

# Calculate total runs scored by each team
total_runs = runs_scored.groupby('Team')['Runs'].sum().reset_index()

# Plot the data
plt.figure(figsize=(12, 6))
sns.barplot(x='Runs', y='Team', data=total_runs.sort_values(by='Runs', ascending=False), palette='inferno')
plt.title('Total Runs Scored by Each Team')
plt.xlabel('Total Runs')
plt.ylabel('Team')
plt.show()
```

```python
# Assuming 't20' is your dataframe containing '1st Team' and '2nd Team' columns
matches_played = pd.concat([t20['1st Team'], t20['2nd Team']])
matches_played_counts = matches_played.value_counts().reset_index()
matches_played_counts.columns = ['Team', 'Matches Played']

plt.figure(figsize=(12, 6))
sns.barplot(x='Matches Played', y='Team', data=matches_played_counts, palette='tab20c')
plt.title('Number of Matches Played by Each Team')
plt.xlabel('Matches Played')
plt.ylabel('Team')
plt.show()
```

```python
# Ensure the 'Economy Rate' and 'Overs' columns are numeric
bowl['Economy Rate'] = pd.to_numeric(bowl['Economy Rate'], errors='coerce')
bowl['Overs'] = pd.to_numeric(bowl['Overs'], errors='coerce')

# Filter out bowlers with the most overs (top 10% most overs)
overs_threshold = bowl['Overs'].quantile(0.90)
filtered_df = bowl[bowl['Overs'] >= overs_threshold]

# Sort and select the top 5 bowlers with the best (lowest) economy rate
top_5_bowlers_best_economy = filtered_df[['Player', 'Economy Rate', 'Overs']].sort_values(by='Economy Rate').head(5)

# Creating the plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Economy Rate', y='Player', data=top_5_bowlers_best_economy, palette='gnuplot2')
plt.title('Top 5 Bowlers with the Best Economy Rate Based on Most Numbers of Overs')
plt.xlabel('Economy Rate')
plt.ylabel('Bowler')
plt.show()
```
